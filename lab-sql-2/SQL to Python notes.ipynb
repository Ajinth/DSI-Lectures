```python
sql = """
SELECT foods.*, restaurants.name
FROM foods
INNER JOIN restaurants ON restaurants.id = foods.restaurant_id
"""
pd.read_sql_query(sql,engine).head()
```

```python
sql = """
SELECT foods.name, foods.calories, foods.carbs, foods.fat, 
restaurants.name AS restaurant, 
categories.name AS category
FROM foods
INNER JOIN restaurants ON restaurants.id = foods.restaurant_id
INNER JOIN categories on categories.id = foods.category_id
"""
pd.read_sql_query(sql,engine).head()
```

```python
sql = """
SELECT foods.name, foods.calories, foods.carbs, foods.fat, restaurants.name as restaurant
FROM foods
INNER JOIN restaurants ON restaurants.id = foods.restaurant_id
WHERE restaurants.name LIKE 'Wendy%%'
ORDER BY foods.calories DESC
"""
pd.read_sql_query(sql, engine).head()
```

```python
sql = """
SELECT foods.name, foods.calories, foods.carbs, foods.fat, restaurants.name as restaurant
FROM foods
INNER JOIN restaurants ON restaurants.id = foods.restaurant_id
INNER JOIN categories ON categories.id = foods.category_id
WHERE categories.name = 'Shakes'
ORDER BY foods.calories DESC
"""
pd.read_sql_query(sql, engine).head(20)
```

```python
sql = """
SELECT categories.name, AVG(foods.calories)
FROM categories
INNER JOIN foods ON categories.id = foods.category_id
GROUP BY categories.name
ORDER BY AVG(foods.calories) ASC
"""
pd.read_sql_query(sql, engine)
```

```python
# give the most caloric food item per restaurant
sql = '''SELECT restaurants.name, max(foods.calories)
FROM restaurants
INNER JOIN foods ON foods.restaurant_id = restaurants.id
GROUP BY restaurants.name
ORDER BY max(foods.calories) DESC
'''

pd.read_sql_query(sql, engine)
```

```python
# Restaurant name, food name, calories from all kids meals
sql = """
SELECT restaurants.name, foods.name as food, foods.calories
from restaurants
INNER JOIN foods ON foods.restaurant_id = restaurants.id
INNER JOIN categories ON categories.id = foods.category_id
WHERE categories.name = 'Kid''s Meals'
ORDER BY foods.calories DESC
"""
pd.read_sql_query(sql,engine).head()
```
